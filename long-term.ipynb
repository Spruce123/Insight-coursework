```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("btc_blocks.csv")
difficulty_list = df.difficulty.values.tolist()
transaction_list = df.transaction_count.values.tolist()
size_list = df.block_size.values.tolist()
usd_list = df.output_total_usd.values.tolist()
total_list = df.output_total.values.tolist()
input_list = df.input_count.values.tolist()
output_list = df.output_count.values.tolist()
price_list = []
address_list = []
p = 0
for i in range(len(usd_list)):
    if(total_list[i]):
        p = usd_list[i]/total_list[i]
    price_list.append(p)
for i in range(len(usd_list)):
    address_list.append(input_list[i] + output_list[i])
plt.subplot(6,1,1)
plt.plot(difficulty_list)
plt.subplot(6,1,2)
plt.plot(transaction_list)
plt.subplot(6,1,3)
plt.plot(size_list)
plt.subplot(6,1,4)
plt.plot(price_list)
plt.subplot(6,1,5)
plt.plot(usd_list)
plt.subplot(6,1,6)
plt.plot(address_list)
```
